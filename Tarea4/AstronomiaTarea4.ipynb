```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#Nombro aqui las constantes que utilizare en el programa
h=6.62606957e-34 #Constante de Planck kg*m^2/s
c=2.99792458e8 #Velocidad de la luz m/s
k=1.38064880e-23 #Constante de Boltzmann J/K
```

```python
#Funci√≥n de Planck para la intensidad del cuerpo negro
#Donde:
#T es la temperatura en Kelvins
#ld es la longitud de onda en angstroms
def planck(ld, T):
    B = (2.0*h*c**2) / ( (ld**5) * (np.exp((h*c)/(ld*k*T)) - 1.0) )
    return B
```

```python
#1000 Kelvins
#Intervalo de evaluacion para la longitud de onda desde 1 a 80000 angstroms
longitudA = np.arange(1e-10, 10e-6, 1e-10)
#Evaluacion
intensidadA = planck(longitudA, 1000.)

#Grafica
plt.figure(figsize=(10,5))
plt.plot(longitudA*1e10, intensidadA)
plt.title('Gr√°fica de cuerpo negro a 1000K')
plt.xlabel('√Ö‚Äã')
plt.ylabel('Intensidad (1e+9)')
plt.grid()
plt.show()
```

**4.- Para un cuerpo de 5770K:
Graficar la ecuacion de Planck y la aproximacion de Rayleigh-Jeans como funcion de longitud de onda**

```python
#Considerando un sistema de unidades mas peque√±as se cambiara el valor de las constantes

longitud2 = np.arange(1e-9, 2e-5, 1e-10)

#c = 3.0e+10  #Velocidad de la luz cm/s
#k = 1.3806488e-16     #En erg‚ÄâK‚àí1

h = 6.626e-34
c = 2.998e+8
k = 1.381e-23

#Aproximaci√≥n de Rayleigh-Jeans
def RJ(ld2, T):
    RJ = 2*c*k*T/ld2**4
    return RJ


planck5770 = planck(longitud2, 5770.)
RJ5770 = RJ(longitud2, 5770.)
plt.figure(figsize=(10,5))
plt.plot(longitud2*1e10, planck5770, 'r')
plt.plot(longitud2*1e10, RJ5770, 'b')
plt.title('Gr√°fica de Planck y Rayleigh-Jeans ')
plt.xlabel('√Ö')
plt.ylabel('Intensidad (1e+14)')
plt.xlim(0,5e4)
plt.ylim(0,10.0e13)
plt.grid()
plt.show()
```
