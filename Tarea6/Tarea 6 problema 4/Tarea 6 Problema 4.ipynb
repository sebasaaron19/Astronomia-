```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#Cargo el archivo de datos con el que trabajare
data = np.loadtxt('NGC6604_phot_corr.dat',skiprows=2, usecols=(0,4,5)) 
#Creo un DataFrame
df=pd.DataFrame(data, columns=['Est-ID','B','V']) 
#Me deshago de los datos que no me sirven
df.drop(df[df['B'] >50.00 ].index , inplace=True)
df.drop(df[df['V'] >50.00 ].index , inplace=True)
#Creo una nueva columna con el color B-V
df['B-V']=df['B']-df['V']
#Reinicio el indice
df=df.reset_index(drop=True)
```

```python
x=df['B-V']
y=df['V']
plt.figure(figsize=[8,4])
ax=sns.scatterplot(x,y,color='r')
plt.title('Diagrama HR corregido de NGC6604')
plt.ylim(max(y)+0.5, min(y)-0.5)
plt.xlabel('Color B-V')
plt.ylabel('Magnitud V (aparente)')
plt.grid(True)
```

**Las isocronas que mejor se ajustan al cúmulo NGC6604 son las de 50, 100 y 200 millones de años**

```python
#Calculo de la edad promedio e incertidumbre
#Promedio
Ed=(200+100+50)/2
#error
ed0=200-Ed
ed1=100-Ed
ed2=50-Ed
#La incertidumbre la caculare con la formula S=sqrt((sumatoria(Xn-Xprom)**2)/n-1)
inc=np.sqrt(((ed0+ed1+ed2)**2)/2)
print('La edad promedio del cúmulo es',Ed,'millones de años')
print('con una incertidumbre de',np.round(inc,4),'millones de años')
```
